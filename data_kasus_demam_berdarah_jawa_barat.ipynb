```python
#data kasus demam berdarah berdasarkan kabupaten/kota di Jawa Barat per tahun
df = pd.read_csv('/content/drive/MyDrive/dinkes-od_17308_jml_kasus_demam_berdarah_dengue_dbd__jk_v1.csv')
sorted(df['tahun'])
df.groupby(['tahun','nama_kabupaten_kota'])['jumlah_kasus'].sum()
```

```python
# Total jumlah kasus demam berdarah di Jawa Barat didominasi oleh laki-laki sebesar ± 118.000 kasus
cases_by_gender = df.groupby('jenis_kelamin')['jumlah_kasus'].sum().reset_index()

# Create the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(data=cases_by_gender, x='jenis_kelamin', y='jumlah_kasus', palette='viridis')
plt.title('Total Jumlah Kasus DBD Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Total Jumlah Kasus')
plt.show()
```

```python
# Tren data kasus demam berdarah tertinggi di daerah Jawa Barat terjadi pada tahun 2016
cases_by_year = df.groupby('tahun')['jumlah_kasus'].sum().reset_index()

# Create the line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=cases_by_year, x='tahun', y='jumlah_kasus', marker='o')
plt.title('Tren Jumlah Kasus DBD per Tahun')
plt.xlabel('Tahun')
plt.ylabel('Total Jumlah Kasus')
plt.grid(True)
plt.xticks(cases_by_year['tahun'])
plt.show()
```

```python
#Tren kasus demam berdarah jika dibandingkan berdasarkan jenis kelamin
cases_by_year_gender = df.groupby(['tahun', 'jenis_kelamin'])['jumlah_kasus'].sum().reset_index()

plt.figure(figsize=(12, 7))
sns.lineplot(data=cases_by_year_gender, x='tahun', y='jumlah_kasus', hue='jenis_kelamin', marker='o')
plt.title('Perbandingan Jumlah Kasus DBD per Tahun Berdasarkan Jenis Kelamin')
plt.xlabel('Tahun')
plt.ylabel('Total Jumlah Kasus')
plt.grid(True)
plt.xticks(cases_by_year_gender['tahun'].unique())
plt.legend(title='Jenis Kelamin')
plt.show()
```
