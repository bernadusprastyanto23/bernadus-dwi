```python
#data kasus berdasarkan tahun di kabupaten/kota
df = pd.read_csv('/content/drive/MyDrive/dinkes-od_17308_jml_kasus_demam_berdarah_dengue_dbd__jk_v1.csv')
sorted(df['tahun'])
df.groupby(['tahun','nama_kabupaten_kota'])['jumlah_kasus'].sum()
```

```python
# Group data by 'jenis_kelamin' and sum 'jumlah_kasus'
cases_by_gender = df.groupby('jenis_kelamin')['jumlah_kasus'].sum().reset_index()

# Create the bar plot
plt.figure(figsize=(8, 6))
sns.barplot(data=cases_by_gender, x='jenis_kelamin', y='jumlah_kasus', palette='viridis')
plt.title('Total Jumlah Kasus DBD Berdasarkan Jenis Kelamin')
plt.xlabel('Jenis Kelamin')
plt.ylabel('Total Jumlah Kasus')
plt.show()
```
