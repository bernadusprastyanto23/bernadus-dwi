# 6. Conditional

```python
# klasifikasi nilai
# nilai>=81 : "A"
# 76<=nilai<81 : "AB"
# nilai <76 : 'C'
nilai = 83

if nilai>=81:
   print('A')
elif nilai>=76 and nilai<81:
    print('AB')
else :
    print('C')

```
